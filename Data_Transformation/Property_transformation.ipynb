```python
import pandas as pd
import re
from googletrans import Translator



translator = Translator()

def translate_text(text):
    if pd.isna(text):  # Check for NaN values
        return 'na'  # Replace NaN with a specific value
    try:
        translation = translator.translate(text, dest='en')
        print(translation.text)
        return translation.text if translation else ''  # Handle None return value
    except Exception as e:
        print(f"Error translating text: {text}. Error: {e}")
        return ''  # Return an empty string in case of error

def clean_text(text):
    if pd.isna(text):  # Check for NaN values
        return ''  # Replace NaN with an empty string
    text = text.lower()  # Lowercase
    text = re.sub(r'[^a-z0-9\s]', '', text)  # Remove special characters
    text = ' '.join(text.split())  # Trim whitespaces
    return text.strip()

# First, translate to English
data['translated_name'] = data['name'].apply(translate_text)

# Then, clean the translated text
data['cleaned_name'] = data['translated_name'].apply(clean_text)

# Print the cleaned names
print(data[['name', 'translated_name', 'cleaned_name']])
```
